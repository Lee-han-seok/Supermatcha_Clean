```python
from sklearn.cluster import KMeans 
def elbow(X) :
    sse = []
    for i in range(1,11) :
        km = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
        km.fit(X)
        sse.append(km.inertia_)
        
    plt.plot(range(1,11), sse, marker = 'o')
    plt.xlabel('n_cluster')
    plt.ylabel('SSE')
    plt.show()
    
elbow(principalDf)
```

```python
def percentage_pivot(df, columns, index, values, aggfunc = len, fill_value = 0):
    f = pd.pivot_table(df, columns = columns, index = index, values = values, aggfunc = aggfunc, fill_value = fill_value, margins = True)
    f2 = f.div(f.iloc[:, -1], axis = 0)
    return f2.iloc[:-1, :-1]
```

```python
percentage_pivot(data, index = 'c_3', columns = '직업', values = '나이',aggfunc = len, fill_value = 0).plot(kind = 'bar', figsize = (12, 6))
plt.title('그룹별 직업 분포')

plt.ylim(0,1)

plt.show()
```
