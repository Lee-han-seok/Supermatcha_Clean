```python
import requests 
from selenium import webdriver as wd 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
import time
import random
import re
import pandas as pd
```

```python
keyword = input()
url = "https://www.instagram.com/explore/tags/{}/".format(keyword)
loginUrl = 'https://www.instagram.com/accounts/login/'
username = '2_hanseok'
userpw = 'qltks1128!?'
```

```python
instagram_post = []
instagram_hashtags = []
instagram_date = []
instagram_likes = []
instagram_locate = []
driver = wd.Chrome('C:/Users/onest/OneDrive/바탕 화면/Python/hitthetea/인스타크롤링_최종/chromedriver.exe')
driver.get(loginUrl)
time.sleep(3)

driver.find_element_by_name('username').send_keys(username)
driver.find_element_by_name('password').send_keys(userpw)
driver.find_element_by_name('password').submit()
time.sleep(5)

driver.get(url) 
time.sleep(3)

driver.find_element_by_css_selector('div.v1Nh3.kIKUG._bz0w').click()

for i in range(10000):
    try:
        time.sleep(random.uniform(2,4))
        #해시태그 수집
        button = driver.find_elements_by_css_selector('button.dCJp8.afkep span')
        if(button != []):
            while True:
                time.sleep(2)
                if(button == []):
                    break
                elif(button != []):
                    button[0].click()
                button = driver.find_elements_by_css_selector('button.dCJp8.afkep span')
            time.sleep(1)       
            button = driver.find_elements_by_css_selector('button.sqdOP.yWX7d.y3zKF span')
            if(button == []):
                hashtag = driver.find_elements_by_css_selector('.xil3i')
                hashtags = [hashtag[i].text.replace("#","") for i in range(0,len(hashtag))]
                instagram_hashtags.append(hashtags)
            elif(button != []):
                for i in range(0,len(button)):
                    button[i].click()
                hashtag = driver.find_elements_by_css_selector('.xil3i')
                hashtags = [hashtag[i].text.replace("#","") for i in range(0,len(hashtag))]
                instagram_hashtags.append(hashtags)
        elif(button == []):
            button = driver.find_elements_by_css_selector('button.sqdOP.yWX7d.y3zKF span')
            if(button == []):
                hashtag = driver.find_elements_by_css_selector('.xil3i')
                hashtags = [hashtag[i].text.replace("#","") for i in range(0,len(hashtag))]
                instagram_hashtags.append(hashtags)
            elif(button != []):
                for i in range(0,len(button)):
                    button[i].click()
                hashtag = driver.find_elements_by_css_selector('.xil3i')
                hashtags = [hashtag[i].text.replace("#","") for i in range(0,len(hashtag))]
                instagram_hashtags.append(hashtags)
        #좋아요 수집
        likes = driver.find_elements_by_css_selector('div.Nm9Fw span')
        if(likes == []):
            instagram_likes.append([])
        elif (likes != []):
            instagram_likes.append([likes[i].text for i in range(0,len(likes))])
        #날짜 수집
        date_raw = driver.find_elements_by_css_selector('time.FH9sR.Nzb55')
        if(date_raw == []):
            instagram_date.append([])
        elif(date_raw != []):
            instagram_date.append(date_raw[0].get_attribute('title'))
        #위치 수집
        locate = driver.find_elements_by_css_selector('.O4GlU')
        if(locate == []):
            instagram_locate.append([])
        elif (locate != []):
            instagram_locate.append([locate[0].text for i in range(0,len(locate))])
        #게시글 수집 
        post = driver.find_elements_by_css_selector("div.C4VMK span")
        if(post == []):
            instagram_post.append([])
        elif(post != []):
            instagram_post.append([post[1].text])
    except:
        instagram_post.append(["error"])
        instagram_hashtags.append(["error"])
        instagram_date.append(["error"])
        instagram_likes.append(["error"])
        instagram_locate.append(["error"])
    try:
        WebDriverWait(driver, 300).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'a.coreSpriteRightPaginationArrow')))
        driver.find_element_by_css_selector('a.coreSpriteRightPaginationArrow').click() 
    except:
        driver.close()    
    time.sleep(random.uniform(2,4))
```

```python
#json 저장
```

```python
df = pd.DataFrame({'Post':instagram_post,
                   'Hashtags':instagram_hashtags,
                   'Locates':instagram_locate,
                   'Date':instagram_date,
                   'Likes':instagram_likes,
                  })
```
