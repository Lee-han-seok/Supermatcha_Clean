```python
df = data[['keyword1', 'keyword2','keyword3','keyword4','keyword5','keyword6','keyword7','keyword8','keyword9']]
```

```python
df
```

```python
df = df.fillna('')

a = []
for j in range(0,9):
    line = []
    for i in df.iloc[:,j]:
        word = i.split(',')
        line.append(word)       
    a.append(line)

keyword = []
for i in range(0,33):
    keyword.append(a[0][i] + a[1][i] + a[2][i] + a[3][i] + a[4][i] + a[5][i] + a[6][i] + a[7][i] + a[8][i])
    
keyword_list = sum(keyword, [])
while '' in keyword_list:
    keyword_list.remove('')

keyword_list
```

```python
keyword_list
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
```

```python
stop_words = [""]
```

```python
cv = CountVectorizer(max_features=100, stop_words = stop_words)
```

```python
tdm = cv.fit_transform(keyword_list)
```

```python
word_count = pd.DataFrame({
    '단어': cv.get_feature_names(),
    '빈도': tdm.sum(axis=0).flat 
})
```

```python
wc = WordCloud(background_color='white',
               width=800, height=800, 
               font_path='NanumBarunGothicLight.ttf',
              colormap = 'hsv', # https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
              max_words = 200)
```

```python
count_dic = word_count.set_index('단어')['빈도'].to_dict()
cloud = wc.fit_words(count_dic)
```

```python
cloud.to_image()
```

```python
tdm.shape
```

```python
# one hot encoding
import numpy as np

one_hot = np.eye(tdm.shape[1])
```

```python
words = cv.get_feature_names()
```

```python
tokens = 'keyword_list'
```

```python
%%time
doc = np.zeros((1, tdm.shape[1]))

for token in tokens:
    try:
        i = words.index(token)
        doc += one_hot[i]
    except ValueError:
        pass
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=36)
```

```python
svd.fit(tdm)
```

```python
svd.components_.shape
```

```python
word_emb = svd.components_.T
```

```python
indices = []
target = list(word_count.sort_values('빈도', ascending=False)['단어'][0:100])
for w in target:
    i = words.index(w)
    indices.append(i)
    print(w, i)
```
