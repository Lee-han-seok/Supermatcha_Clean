```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('힛더티 클린 오프라인 조사_수정1.csv', encoding = 'cp949') ; df
```

```python
df = df.fillna('')

a = []
for j in range(0,9):
    line = []
    for i in df.iloc[:,j+12]:
        word = i.split(',')
        line.append(word)       
    a.append(line)

keyword = []
for i in range(0,33):
    keyword.append(a[0][i] + a[1][i] + a[2][i] + a[3][i] + a[4][i] + a[5][i] + a[6][i] + a[7][i] + a[8][i])

keyword
```

```python
for i in keyword:
    while '' in i:
        i.remove('')

keyword
```
